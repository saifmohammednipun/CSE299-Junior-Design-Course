```python
# Calculate gender distribution
gender_distribution = diabetes['gender'].value_counts(normalize=True) * 100
gender_distribution
```

```python
# Plot histograms for numerical columns
diabetes.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
# Plot count of gender
plt.figure(figsize=(6, 4))
sns.countplot(x='gender', data=diabetes)
plt.title("Distribution of Gender")
plt.show()

# Plot count of diabetic
plt.figure(figsize=(6, 4))
sns.countplot(x='diabetic', data=diabetes)
plt.title("Distribution of Diabetic")
plt.show()
```

```python
import math

# List of numerical features
numerical_features = [col for col in diabetes.columns if diabetes[col].dtype in ['int64', 'float64']]

# Calculate the number of rows and columns needed
num_features = len(numerical_features)
num_cols = 3  # Number of columns in the subplot grid
num_rows = math.ceil(num_features / num_cols)  # Number of rows needed
```

```python
plt.figure(figsize=(15, 5 * num_rows))  # Adjust figure size based on the number of rows

for i, feature in enumerate(numerical_features, 1):
    plt.subplot(num_rows, num_cols, i)  # Use dynamic grid size
    sns.boxplot(x='diabetic', y=feature, data=diabetes)
    plt.title(f"{feature} vs Diabetic")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='gender', hue='diabetic', data=diabetes)
plt.title("Gender vs Diabetic")
plt.show()
```

```python
# Encode 'gender' and 'diabetic' columns
diabetes['gender'] = diabetes['gender'].map({'Male': 1, 'Female': 0})
diabetes['diabetic'] = diabetes['diabetic'].map({'Yes': 1, 'No': 0})
```

```python
diabetes['bmi_category'] = pd.cut(
    diabetes['bmi'], 
    bins=[0, 18.5, 24.9, 29.9, float('inf')],  # Define BMI bins
    labels=[1, 2, 3, 4],  # Assign numerical labels
    right=True  # Include the right edge of the bin
)
```

```python
diabetes['bp_ratio'] = diabetes['systolic_bp'] / diabetes['diastolic_bp']
```

```python
# Compute correlation matrix
corr_matrix = diabetes.corr()

# Display the correlation matrix
corr_matrix
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
